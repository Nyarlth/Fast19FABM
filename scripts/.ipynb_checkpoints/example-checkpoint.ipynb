```python
from Zfitting import *
import h5py as h5
```

# Load FA data

```python
with h5.File("../data/19FA.h5",'r') as f:
    print(f.keys())
    print(f.attrs.keys())
    B = f["Origin"][:]
    Zf = f["Fitted"][:]
    coef = f["Coefficient"][:]
    Z_order = f.attrs["Z_order"]
    xx = f.attrs["X_coor"][:]
    yy = f.attrs["Y_coor"][:]
    fig_size = f.attrs["fig_size"]
    polar = f.attrs["polar"]
```

# Plot Beam_figure

```python
Beam = 1 -1
```

## stack origin 

```python
plt.pcolormesh(xx, yy, np.ma.log10(B[Beam]), vmax=0, vmin=-2.5, cmap= 'jet')
plt.gca().set_aspect('equal')
plt.ylabel(f"$\delta V \ [arcmin]$")
plt.xlabel(f"$\delta H \ [arcmin]$")
```

## Zernike fitting full coef

```python
plt.pcolormesh(xx, yy, np.ma.log10(Zf[Beam]), vmax=0, vmin=-2.5, cmap= 'jet')
plt.gca().set_aspect('equal')
plt.ylabel(f"$\delta V \ [arcmin]$")
plt.xlabel(f"$\delta H \ [arcmin]$")
```

## Zernike fitting de-noise result

```python
cart = RZern(Z_order)
L,K = len(xx)-1, len(yy)-1
ddx = np.linspace(-1.0 + 1/K, 1.0 - 1/K, K)
ddy = np.linspace(-1.0 + 1/L, 1.0 - 1/L, L)
xv, yv = np.meshgrid(ddx, ddy)
cart.make_cart_grid(xv, yv)
```

```python
c_ = coef[Beam].copy()
mx_ = max(abs(c_))
c_[abs(c_) <mx_*0.01] = 0
Phi = cart.eval_grid(c_, matrix = True)
```

```python
plt.pcolormesh(xx, yy, np.ma.log10(Phi), vmax=0, vmin=-2.5, cmap= 'jet')
plt.gca().set_aspect('equal')
plt.ylabel(f"$\delta V \ [arcmin]$")
plt.xlabel(f"$\delta H \ [arcmin]$")
```

# Sky Project

```python
# sources status
s_ra   = np.array([129.9, 130.2])
s_dec  = np.array([26.3, 26.3])
s_flux = np.array([10,20])  # mJy
# scan path 
scan_dec = np.array([26.27])
scan_ra = 130.001 + np.linspace(-0.3,0.3,100)
```

```python
Beam = 8-1
c_ = coef[Beam].copy()
mx_ = max(abs(c_))
c_[abs(c_) <mx_*0.01] = 0
cart = Beam_cart(Z_order=Z_order, size = fig_size)
```

```python
vis = mapping((s_ra,s_dec,s_flux), (scan_ra,scan_dec), c_, cart = cart, rot = 23.4)
```

```python
plt.plot(scan_ra, vis)
plt.xlabel(f"$RA. [^\circ]$")
plt.ylabel("Flux [mJy]")
```
